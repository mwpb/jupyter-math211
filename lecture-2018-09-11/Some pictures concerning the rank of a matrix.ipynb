Consider the homogeneous system associated to the matrix

$$\left[\begin{array}{rr}
1 & 2\\
3 & 4\\
12 & 20
\end{array}\right]$$

```python
A = [
    [1, 2, 0],
    [3, 4, 0],
    [12, 20, 0]
]

plot2DSystem(A)
steps = rref(A)
for i in steps:
    printM(i)
plot2DSystem(A)
```

The first graph shows that the solution to the system is at the intersection of three lines. (In this case just the point (0,0).) However we only need two lines to actually specify this point. The rank of a matrix is the minimum number that we need to specify all of the solutions to the system. In this case the rank is 2.

Note that in the process of performing Gaussian elimination the superfluous lines are removed. (They become the zero rows at the bottom of the associated reduced row echelon matrix.) The second graph shows the graphical interpretation of the reduced row echelon form.
